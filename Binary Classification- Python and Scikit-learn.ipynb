# Binary Classification normal pythonic way and scikit-learn method

Several regression and binary classification algorithms are available in scikit-learn. In this part I am first
classifying the quality of wine using map function (threshold) of python then I used LabelBinarizer preprocessing
function of scikit-learn method to classify wine either 0 or 1. 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/Users/sundarpandey/Documents/Analyticalvidhya/winne.xlsx', skiprows=1)
df
```

```python
df.columns
```

```python
df1 = df[['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'Alcohol',
       'quality', 'Value']]
df1
```

# Pythonic way
Here 'Yes' could be replace with 1 and 'No' with 0

```python
df1['quality']= df1['quality'].map(lambda x:  'Yes' if x <= 5 else 'No')
df1
```

```python
df1['quality']
```

# Scikit-learn way

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
```

```python
lb.fit(df1['quality'])
```

```python
df1['quality']= lb.transform(df1['quality'])
df1
```
